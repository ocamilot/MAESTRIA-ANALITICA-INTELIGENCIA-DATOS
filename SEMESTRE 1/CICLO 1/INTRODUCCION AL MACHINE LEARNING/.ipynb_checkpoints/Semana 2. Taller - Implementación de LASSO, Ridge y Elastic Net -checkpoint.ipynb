### Penalizacin en modelos lineales

Adicional a la seleccin de variables en un modelo y a la transformacin de variables usando PCA o PLS, los modelos lineales con penalizacin son una tercera alternativa para calibrar mejor y ms fcil los modelos de regresin lineal.

Recordemos que inicialmente tenemos el modelo de mnimos cuadrados como punto de partida. As, este modelo se estima resolviendo el siguiente problema de optimizacin:

$$\hat\beta = argmin \sum_{i=1}^{n} (Y_i-\beta^tX_i)^2$$

Sin embargo, es posible que la redundancia de informacin entre las variables predictoras genere que la estimacin de los betas tenga mucha varianza y, por lo tanto, algunos betas tomen valores absolutos muy grandes.

Por tal motivo, los modelos que se van a presentar en este notebook penalizan el tamao del vector de betas para evitar el anterior inconveniente, tal y como se presenta en el siguiente problema de optimizacin:

$$\hat\beta = argmin \sum_{i=1}^{n} (Yi-\beta^tX_i)^2 + \lambda ||\beta|| $$

Donde $\lambda>0$ es el parmetro de penalizacin y $||\beta||$ corresponde a una norma del vector de betas.

La diferencia entre los tres modelos de penalizacin que se presentarn a continuacin, consiste en la manera de definir la norma del vector de betas.

### Libreras a importar

A continuacin se muestran los comandos a ejecutar para poder implementar los procedimientos de este notebook:

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNetCV
from sklearn.linear_model import ElasticNet
import matplotlib.pyplot as plt
```

### Cargando los datos

En primer lugar, cargaremos los datos a usar en este Notebook que corresponde a la base de datos "Diabetes" del paquete "sklearn". Para ello, ejecutaremos el siguiente cdigo:

```python
XTotal, yTotal = datasets.load_diabetes(return_X_y=True)
```

Luego, partiremos la base de datos en una seccin de train y otra de test de la siguiente manera: utilizaremos el 67% de los datos para entrenamiento y el 33% restante para evaluacin.

```python
XTrain, XTest, yTrain, yTest = train_test_split(XTotal, yTotal, test_size=0.33, random_state=0)
```

Ahora, procederemos a discutir las tres modelos de penalizacin discutidos anteriormente y los cdigos para implementarlos.

### Modelo con penalizacin tipo Ridge

Este modelo utiliza la norma del vector de betas como la suma del cuadrado de sus componentes, es decir, la norma euclidiana clsica de un vector. De esta manera, el problema de optimizacin que se busca solucionar es:

$$\hat\beta = argmin \sum_{i=1}^{n} (Y_i-\beta^tX_i)^2 + \lambda \sum_{j=1}^{k} \beta_j^{2}$$

A continuacin se muestra el cdigo para calibrar el modelo con penalizacin Ridge usando cross-validation:

```python
n_alphas = 200
alphasCalibrar = np.logspace(-10, 2, n_alphas)
modeloRidge = RidgeCV(alphas=alphasCalibrar,store_cv_values=True).fit(XTrain, yTrain)
```

Una vez calibrado el modelo, podemos saber el valor de $\lambda$, del $R^{2}$ de la regresin y de los coeficientes estimados utilizando los siguientes comandos:

```python
print('El parmetro de penalizacin ptimo es:')
print(modeloRidge.alpha_)
print('El valor del R^2 es:')
print(modeloRidge.score(XTrain, yTrain))
print('Los coeficientes estimados son:')
print(modeloRidge.coef_)
```

Para observar lo que sucede con los coeficientes a medida que cambia el valor del parmetro de penalizacin, se debe correr el siguiente cdigo:

```python
coefs = []
for a in alphasCalibrar:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(XTrain, yTrain)
    coefs.append(ridge.coef_)

# Presentacin de resultados

ax = plt.gca()

ax.plot(alphasCalibrar, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  
plt.xlabel('Parmetro de penalizacin')
plt.ylabel('Coeficientes')
plt.title('Penalizacin Ridge')
plt.axis('tight')
plt.show()
```

Finalmente, podemos hacer predicciones sobre la muestra de evaluacin y calcular el MSE correspondiente con el comando que se muestra enseguida:

```python
prediccionRidge=modeloRidge.predict(XTest)
MSERidge=np.average(np.square(prediccionRidge-yTest))
print('MSE regresin con penalizacin tipo Ridge:')
print(MSERidge)
```

### Modelo con penalizacin tipo LASSO

Este modelo utiliza la norma del vector de betas como la suma del valor absoluto de sus componentes, es decir, el problema de optimizacin que se busca solucionar es:

$$\hat\beta = argmin \sum_{i=1}^{n} (Y_i-\beta^tX_i)^2 + \lambda \sum_{j=1}^{k} |\beta_j|$$

A continuacin se muestra el cdigo para calibrar el modelo con penalizacin LASSO usando cross-validation:

```python
n_alphas = 200
alphasCalibrar = np.logspace(-10, 2, n_alphas)
modeloLASSO = LassoCV(alphas=alphasCalibrar).fit(XTrain, yTrain)
```

Una vez calibrado el modelo, podemos saber el valor de $\lambda$, del $R^{2}$ de la regresin y de los coeficientes estimados utilizando los siguientes comandos:

```python
print('El parmetro de penalizacin es:')
print(modeloLASSO.alpha_)
print('El valor del R^2 es:')
print(modeloLASSO.score(XTrain, yTrain))
print('Los coeficientes estimados son:')
print(modeloLASSO.coef_)
```

Para observar lo que sucede con los coeficientes a medida que cambia el valor del parmetro de penalizacin, se debe correr el siguiente cdigo:

```python
coefs = []
for a in alphasCalibrar:
    lasso = Lasso(alpha=a, fit_intercept=False,max_iter=10000)
    lasso.fit(XTrain, yTrain)
    coefs.append(lasso.coef_)

# Presentacin de resultados

ax = plt.gca()

ax.plot(alphasCalibrar, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('Parmetro de penalizacin')
plt.ylabel('Coeficientes')
plt.title('Penalizacin LASSO')
plt.axis('tight')
plt.show()
```

Finalmente, podemos hacer predicciones sobre la muestra de evaluacin y calcular el MSE correspondiente con el comando que se muestra enseguida:

```python
prediccionLASSO=modeloLASSO.predict(XTest)
MSELASSO=np.average(np.square(prediccionLASSO-yTest))
print('MSE regresin con penalizacin tipo LASSO:')
print(MSELASSO)
```

### Modelo con penalizacin tipo Elastic Net

Este modelo combina la norma euclidiana clsica del vector de betas junto con la norma del vector de betas como la suma del valor absoluto de sus componentes, es decir, es una combinacin entre la penalizacin tipo Ridge y la penalizacin tipo LASSO.
De esta manera, el problema de optimizacin que se busca solucionar es:

$$\hat\beta = argmin \sum_{i=1}^{n} (Y_{i}-\beta^tX_i)^2 + \lambda((1-\alpha) \sum_{j=1}^{k} \beta_j^{2}+ \alpha \sum_{j=1}^{k} |\beta_j|)$$

Como se puede observar, para este tipo de penalizacin hay dos parmetros de calibracin $\lambda$ y $\alpha$. Adems, es importante notar que si tenemos $\alpha=0$ estaramos hablando de un modelo con penalizacin tipo Ridge, y si tenemos $\alpha=1$ estaramos hablando de un modelo con penalizacin tipo LASSO.

A continuacin se muestra el cdigo para calibrar el modelo con penalizacin Elastic Net usando cross-validation:

```python
n_alphas = 200
alphasCalibrar = np.logspace(-10, 2, n_alphas)
modeloElastic = ElasticNetCV(alphas=alphasCalibrar,cv=5, random_state=0).fit(XTrain,yTrain)
```

Una vez calibrado el modelo, podemos saber el valor de $\lambda$, del $R^{2}$ de la regresin y de los coeficientes estimados utilizando los siguientes comandos:

```python
print('El parmetro de penalizacin es:')
print(modeloElastic.alpha_)
print('El valor del R^2 es:')
print(modeloElastic.score(XTrain, yTrain))
print('Los coeficientes estimados son:')
print(modeloElastic.coef_)
```

Para observar lo que sucede con los coeficientes a medida que cambia el valor del parmetro de penalizacin, se debe correr el siguiente cdigo:

```python
coefs = []
for a in alphasCalibrar:
    elasticNet = ElasticNet(alpha=a, fit_intercept=False,max_iter=10000)
    elasticNet.fit(XTrain, yTrain)
    coefs.append(elasticNet.coef_)

# Presentacin de resultados

ax = plt.gca()

ax.plot(alphasCalibrar, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('Parmetro de penalizacin')
plt.ylabel('Coeficientes')
plt.title('Penalizacin Elastic Net')
plt.axis('tight')
plt.show()
```

Finalmente, podemos hacer predicciones sobre la muestra de evaluacin y calcular el MSE correspondiente con el comando que se muestra enseguida:

```python
prediccionElastic=modeloElastic.predict(XTest)
MSEElastic=np.average(np.square(prediccionElastic-yTest))
print('MSE regresin con penalizacin tipo Elastic Net:')
print(MSEElastic)
```

### Comparacin de modelos

Finalmente, se presenta un cdigo que grafica conjuntamente los MSE en test de los tres modelos.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Comparacin de modelos")
ejeX = ['LASSO', 'Ridge','Elastic Net']
ejeY = [MSELASSO,MSERidge,MSEElastic]
ax.bar(ejeX,ejeY)
def addlabels(x,y,plotP):
    for i in range(len(x)):
        plotP.text(i,y[i],y[i])
addlabels(ejeX,ejeY,plt)
plt.show()
```

Como se puede observar, para este caso particular, el desempeo del modelo Elastic Net fue mejor que el de Ridge y LASSO.

### Implementacin de cdigo

A continuacin deber escribir una funcin que reciba como parmetros la matriz de predictores de entrenamiento, la variable de respuesta de entrenamiento, la matriz de predictores de evaluacin, la variable de respuesta de evaluacin y un parmetro llamado "penalizacin" que sea 1 si se quiere estimar un modelo con penalizacin tipo LASSO, 2 si se quiere estimar un modelo con penalizacin tipo Ridge y 3 si se quiere estimar un modelo con penalizacin tipo Elastic Net. Luego, se deber estimar el modelo solicitado con los datos de entrenamiento y calcular las predicciones para el conjunto de evaluacin. Finalmente, se debera retornar el MSE del modelo en test.

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import matplotlib.pyplot as plt

def funcion(XTrain,yTrain,XTest,yTest,penalizacion):
    
    if penalizacion == 1:
        #Penalizacion tipo LASSO
        m = Lasso()
        regresion = "LASSO"
    elif penalizacion == 2:
        #Penalizacion tipo Ridge
        m = Ridge()
        regresion = "Ridge"
    elif penalizacion == 3:
        #Penalizacion tipo ElasticNet
        m = ElasticNet()
        regresion = "ElasticNet"
    else:
        raise ValueError ("Tipo de penalizacion mal seleccionado.  Escriba 1 si se quiere estimar un modelo con penalizacin tipo LASSO, 2 si se quiere estimar un modelo con penalizacin tipo Ridge y 3 si se quiere estimar un modelo con penalizacin tipo Elastic Net")

    #Ajustar modelo con XTrain y  yTrain
    Trainresult = m.fit(XTrain, yTrain)

    #Calculo de prediccion del modelo con xTest
    pred = Trainresult.predict(XTest)

    #Calculo MSE
    MSE = np.average(np.square(pred-yTest))


    print('El valor del R^2 es:')
    print(Trainresult.score(XTrain, yTrain))
    print('Los coeficientes estimados son:')
    print(Trainresult.coef_)
    return ("El valor del MSE con regresion tipo " + regresion + "es: " + str(MSE))

funcion(XTrain,yTrain,XTest,yTest,2)
        
```
