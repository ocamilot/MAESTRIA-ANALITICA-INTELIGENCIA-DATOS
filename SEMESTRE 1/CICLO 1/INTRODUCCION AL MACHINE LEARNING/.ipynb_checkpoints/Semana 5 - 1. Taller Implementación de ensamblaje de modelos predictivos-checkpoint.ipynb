### Rotaciones en $R^2$

Las rotaciones son un tipo particular de transformacin lineal que, al aplicarse sobre un espacio, cambia la orientacin de un vector preservando su longitud. En particular, si se quiere rotar un vector en un ngulo $\theta$, la matriz $A(\theta)$ que representa esta transformacin lineal es definida de la siguiente forma:

$$A=\begin{pmatrix}
Cos(\theta) & -Sen(\theta) \\
Sin(\theta) & Cos(\theta) 
\end{pmatrix}$$

De esta manera, si se quisiera rotar 90 grados el vector $v=\begin{pmatrix}
1 \\
1  
\end{pmatrix}$, entonces esto correspondera a multiplicarlo por la matriz $A(\frac{\pi}{2})$ de la siguiente forma:

$$A(\frac{\pi}{2})v=\begin{pmatrix}
Cos(\frac{\pi}{2}) & -Sen(\frac{\pi}{2}) \\
Sin(\frac{\pi}{2}) & Cos(\frac{\pi}{2}) 
\end{pmatrix} \begin{pmatrix}
1 \\
1  
\end{pmatrix}= \begin{pmatrix}
0 & -1 \\
1 & 0 
\end{pmatrix}\begin{pmatrix}
1 \\
1  
\end{pmatrix}=\begin{pmatrix}
-1 \\
1  
\end{pmatrix}$$

A continuacin se muestra el cdigo que realiza esta operacin y la representacin grfica:

```python
vectorOriginal=[1,1] #En rojo

theta = np.radians(90)
A = np.array(((np.cos(theta), -np.sin(theta)), (np.sin(theta), np.cos(theta))))

vectorRotado=np.matmul(A, vectorOriginal) #En azul
print(vectorRotado)
V = np.array([vectorOriginal, vectorRotado])
origin = np.array([[0, 0],[0, 0]])

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b'],scale=5)
plt.show()
```

### Generacin de nmeros aleatorios

Ahora bien, para el cdigo que debern desarrollar al final de este archivo, ser til la implementacin de un generador de nmeros aleatorios (RNG por sus siglas en ingls).

En particular se necesitarn dos tipos de RNG: uno para una variable aleatoria uniforme discreta y otra para una variable aleatoria uniforme continua.

En primer lugar, el cdigo para generar $nRNG$ cantidad de variables aleatorias uniformes continuas entre $a$ y $b$ con $a<b$ es:

```python
a=0
b=1
nRNG=10
np.random.uniform(a,b,nRNG)
```

Ahora bien, si se quiere generar $nRNG$ cantidad de variables aleatorias uniformes discretas entre $a$ y $b$ con $a<b$, entonces deber usarse el siguiente cdigo:

```python
a=0
b=10
nRNG=10
np.round_(np.random.uniform(a,b,nRNG),0)
```

### Implementacin de cdigo para Rotation Bagged Trees

A continuacin deber escribir una funcin que implemente Rotation Bagged Trees. Para ello, deber hacer que la funcin reciba como parmetros la matriz de predictores de entrenamiento, la variable de respuesta de entrenamiento, la matriz de predictores de evaluacin, la variable de respuesta de evaluacin y un parmetro llamado "numArboles" que corresponde al nmero de rboles que se deben utilizar en el ensamblaje.
El cdigo debe entrenar la cantidad de rboles de decisin especificada, utilizando en cada uno de ellos nicamente dos predictores escogidos aleatoriamente. Una vez han sido escogidos los dos predictores a utilizar para cada rbol, se deber escoger, tambin para cada rbol, un ngulo de manera aleatoria entre 0 y $2\pi$ para rotar los predictores y estimar el modelo con los predictores rotados.
Posteriormente, se deber realizar la prediccin en test y promediar los resultados de todos los rboles para obtener la salida  del ensamblaje.
Finalmente, la funcin deber arrojar el MSE de la muestra de test.

```python
def funcion(XTrain,yTrain,XTest,yTest,numArboles):

    def bootstrapping (x, y):
        a=0
        b=len(x)-1
        nRNG=len(x)
        indices_x= np.round_(np.random.uniform(a,b,nRNG),0)
        indices_x= indices_x.astype(int)
        return x[indices_x], y[indices_x]

    def dos_variables_aleatorias(x):
        a=0
        b=x.shape[1]-1
        nRNG=2
        indices_x= np.round_(np.random.uniform(a,b,nRNG),0)
        indices_x= indices_x.astype(int)
        return x[:,indices_x],indices_x
                
  
    def angulo_aleatorio():
        a=0
        b=2*np.pi
        nRNG=1
        radianes=np.random.uniform(a,b,nRNG)[0]
        return radianes  
        
    def rotacion(x_train, theta):
        vectorRotado=np.empty((0, 2))
        A = np.array(((np.cos(theta), -np.sin(theta)), (np.sin(theta), np.cos(theta))))
        for vect in x_train:
            vectorOriginal=vect
            vectorRotado=np.vstack([vectorRotado, np.matmul(A, vectorOriginal)])
        return vectorRotado 
        
    #Iteracion de Arboles
    i=0
    for iteracion in range(numArboles):
        
        #Bootstrapping
        X_train_boot, y_train_boot = bootstrapping(XTrain,yTrain)


        #Seleccion dos predictores aleatorios
        X_train_dos_var, indices_colum_pred = dos_variables_aleatorias(X_train_boot)
        X_test_dos_var=XTest[:,indices_colum_pred]

        #Rotacion
        theta = angulo_aleatorio()
        X_train_rotado= rotacion(X_train_dos_var, theta)
        X_test_rotado= rotacion(X_test_dos_var, theta)
        
        #Arbol de regresion
        arbol = DecisionTreeRegressor()
        arbol  = arbol.fit(X_train_rotado, y_train_boot)
        prediccion=arbol.predict(X_test_rotado)
        if i==0:
            suma_predic = prediccion
        else: 
            suma_predic+=prediccion
        i+=1
    
    #Calculo promedio de las predicciones de los n arboles
    promedio_final_arboles = suma_predic/numArboles
    # Calcular el promedio del error cuadrtico medio
    MSEArboles=np.average(np.square(promedio_final_arboles-yTest))
    print("El MSE del modelo en la muestra de evaluacin es igual a:")
    print(MSEArboles)    


funcion(XTrain,yTrain,XTest,yTest,30)
```
