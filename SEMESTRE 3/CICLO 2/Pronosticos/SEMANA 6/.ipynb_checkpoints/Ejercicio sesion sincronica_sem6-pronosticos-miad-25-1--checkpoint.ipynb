###  Importacion de datos

```r
start<-format(as.Date("2017-05-01"),"%Y-%m-%d")
end<-format(as.Date("2021-12-31"),"%Y-%m-%d")

#--------- Funcisn para bajar precios y generar rendimientos:
rend<-function(simbolo) {
  ##---------Obtener precios de yahoo finance:
  datos<-getSymbols(simbolo,  from=start, to= end, auto.assign = FALSE)
  ##---------eliminar datos faltantes:
  datos<-na.omit(datos)
  ##--------Mantener el precio de interis:
  datos<-datos[,4]
  ##--------Rendimientos simples:
  #rend<-periodReturn(datos, period = "daily", type='arithmetic')                         
  #------ --Para hacer dtos accesibles  GLobal ENv:
  assign(simbolo, datos, envir = .GlobalEnv)
}
```

#### Llamar la funcion para cada activo particular:

```r
rend("MSFT")
rend("INTC")
rend("GOOGL")
```

## VAR y Prueba de Granger:
##### Ref interés: https://www.researchgate.net/post/Is-it-necessary-to-ensure-stationarity-of-all-time-series-variables-when-you-run-a-Vector-Autoregressive-VAR-Model 
# prueba Gragner:
#### H0: La serie x no es gragner-cause de la serie y.
#### H1: La serie x es Gragner-Cause de la serie y.

###### Término Gragner-cause: el valor de la serie x en un cierto rezago, es útil para 
###### predecir el valor de la serie y.
###### Es una prueba F, si el pvalor < alpha, rechazamos la nula, concluyendo que 
###### hay evidencia para decir que la serie x Gragner-Cause la serie y.


### VAR:

```r
library(vars)
```

```r
msft=ts(datos$MSFT, start= c(2019,01,01), frequency = 365)
intc=ts(datos$INTC, start= c(2019,01,01), frequency = 365)
goog= ts(datos$GOOGL, start= c(2019,01,01), frequency = 365)
```

```r
dataVAR2=ts.intersect(msft, intc, goog) %>% na.omit()
head(dataVAR2)
```

<a id="9"></a> <br>
### Var summary

```r
m1 <- vars::VAR(y=dataVAR2, lag.max = 7)
summary(m1)
```

<a id="9a"></a> <br>
### Modelo VAR con MTS y revision de estacionariedad en VAR

```r
#Para ver si es correcto este enfoque, de desarrollar el VAR en portafolio no diferenciado, podemos aplicar la descomposición singular a la matriz de coeficientes regresivos y observar que no haya valores caracteríticos mayores a la unidad.
library(MTS)
m2=VARorder(dataVAR2)
```

```r
#Vemos que el orden regresivo es similar, con l=2. Procedemos con el modelo VAR mediante librería MTS
m2_mts=MTS::VAR(dataVAR2, 2)
```

```r
#Procedemos con el análisis de estacionariedad en la matriz regresiva
str(m2_mts)
```

```r
phi1=m2_mts$Phi[,c(1,2,3)]
phi2=m2_mts$Phi[,c(4,5,6)]
phi1
phi2
```

```r
eigen(phi1)
eigen(phi2)
```

Esta solución pudiera ser útil, sobre todo si no deseamos sacrificar información al diferenciar. Sin embargo, otro método implica el trabajar con portafolios diferenciados, como se abordará más adelante.

<a id="9b"></a> <br>
### Modelo VAR refinado

Por otro lado, mediante la librería de MTS, es posible hacer un modelo depurado respecto aquellos términos no relevantes del anterior modelo, Esto lo hacemos con la función refVAR(modelo_anterior,thres = 1.96)

```r
m3_mts=refVAR(m2_mts,thres = 1.96)
```

```r
#Modelo check:
resi3=m3_mts$residuals
mq(resi3, adj=15) #Ajustamos los grados de libertad
```

```r
#Y revisamos el efecto instantáneo entre las variables:
MTSdiag(m3_mts, adj=12) 
```

Como podemos ver los p-valores en CCM at lag:  0 , no se tiene ningún efecto intantáneo significativo entre las variables.

```r
#Podemos visualizar los pronósticos a través de la función VARpred().
library(forecast)
pred2=VARpred(m3_mts,  h=12)
MTSplot(pred2$pred)
```

<a id="10"></a> <br>
### VAR con portafolio diferenciado:

```r
#Comenzamos diferenciando los activos para luego, unirlos.
msft.d=diff.xts(s1$MSFT.Close)[-1,]
intc.d=diff.xts(s2$INTC.Close)[-1,]
goog.d= diff.xts(i3$GOOGL.Close)[-1,]
adf.test(msft.d)
adf.test(intc.d)
adf.test(goog.d)
```

```r
#Pasamos a formato ts
msft.d.ts=ts(msft.d, start= c(2019,01,01), frequency = 365)
intc.d.ts= ts(intc.d, start= c(2019,01,01), frequency = 365)
goog.d.ts=ts(goog.d, start= c(2019,01,01), frequency = 365)
```

```r
# Y unimos:
dataVAR3=ts.intersect(msft.d.ts, intc.d.ts, goog.d.ts) %>% na.omit()
head(dataVAR3)
```

<a id="11"></a> <br>
### Nivel K Selection

```r
#Nuevamente, seleccionamos nivel autorregresivo:
nivelk3=VARselect(dataVAR3, lag.max = 7, type = "const")
nivelk3$selection
```

<a id="12"></a> <br>
### Johansen Test

```r
#Podemos aplicar prueba Johansen test para verificar si hay estabilidad a largo plazo.
johatest=ca.jo(dataVAR3, type = "trace", K=2, ecdet ="none", spec = "longrun") #K is the number of lags to use in the vector autoregressive model and is set this to the minimum, K=2.
summary(johatest)
```

<a id="13"></a> <br>
### VAR

```r
#Armamos modelo VAR
var_est3 <- vars::VAR(y=dataVAR3, p=2)
summary(var_est3)
```

```r
dataVAR3_f=predict(var_est3, n.ahead=12)
```

<a id="14"></a> <br>
## Recover level forecast

```r
#Como diferenciamos, recuperamos el nivel.
nhor=12 #pasos en el pronóstico.
nr_lev <- nrow(datos)
mr_lev= as.matrix(datos)
str(mr_lev)
m.varf_lev_ft <- rbind(mr_lev, matrix(NA, nhor,3 ))
m.ft_df <- do.call(cbind,lapply(dataVAR3_f$fcst, 
                                function(x) x[,"fcst"]))
```

<a id="15"></a> <br>
### Growth to level

```r
for(h in (nr_lev+1):(nr_lev+nhor)) {
  hf <- h - nr_lev
  m.varf_lev_ft[h,] <- m.varf_lev_ft[h-1,] + m.ft_df[hf,]
}
```

<a id="16"></a> <br>
### Draw Graph

```r
str.main=c("MSFT", "INTC", "GOOGL")
par(mfrow=c(3,1), mar=c(2,2,2,2))

for(i in 1:4) {
  df <- m.varf_lev_ft[,i]
  matplot(df, type=c("l"), col = c("blue"), 
          main = str.main[i]) 
  abline(v=nr_lev, col="red")
}
```
