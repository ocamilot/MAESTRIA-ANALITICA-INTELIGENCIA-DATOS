## Vectores y Matrices
Los _vectores_ son arreglos unidimensionales de nmeros, variables o trminos. Geomtricamente, contienen informacin sobre la magnitud y direccin de un cambio potencial en un punto. Cuando un vector posee ms de una dimensin, se le conoce como una _matriz_. Las matrices ms comunes son de dos dimensiones y consisten en un arreglo rectangular de datos, similar a las hojas de clculo (_spreadsheets_). Entre las formas comunes de representar vectores y matrices, podemos notar las siguientes:
$$

$$

### Dimensiones
Las dimensiones se describen en trminos del nmero de filas y columnas que posee una matriz. Esta definicin tambin se extiende a los vectores, que se pueden entender como matrices de una sola fila y varias columnas, o una sola columna y varias filas. Es importante tener claras las dimensiones de una matriz para poder realizar operaciones consistentes entre arreglos de datos. Para conocer la dimensin de un arreglo en _numpy_, se puede usar la extensin _.shape_ despus del arreglo, el cual nos dar un vector con las dimensiones del arreglo.

```python
a = np.array([
 [1,2,3]
])
print("a =", a)
print("Dimensiones de a =", a.shape)

b = np.array([
 [1],
 [2],
 [3]
])
print("b =\n", b)
print("Dimensiones de b =", b.shape)

C = np.array([
 [1,2,3],
 [4,5,6]
])
print("C =\n", C)
print("Dimensiones de C =", C.shape)
```

Ntese que el vector $a=[1,2,3]$ tiene dimensiones $(1,3)$, es decir, una fila y tres columnas. El vector $b$ contiene los mismos elementos pero es vertical y tiene tres filas y una columna. Finalmente, la matriz $C$ posee dos filas y tres columnas.

### Operaciones con escalares
Estas operaciones involucran un arreglo de datos y un nmero el cual puede sumar, restar, multiplicar o dividir cada uno de los elementos del arreglo. A continuacin se ejemplifican varias operaciones para un vector $a$ y una matriz $B$:

```python
a = np.array([
 [1,2,3]
])
print("a + 1 =\n", a + 1)
print("a*3 =\n", a*3)

B = np.array([
 [1,2,3],
 [4,5,6]
])
print("B - 2 =\n", B - 2)
print("B/2 =\n", B/2)
```

### Operaciones elemento por elemento
En este tipo de operaciones, los valores de cada posicin de un arreglo se suman, restan o multiplican con los de la misma posicin en el otro arreglo. Esto implica que, tericamente, los vectores y matrices a operar deben tener dimensiones iguales.

#### Adicin y sustraccin
$$
\begin{bmatrix}
a_1 & a_2\\
a_3 & a_4
\end{bmatrix}
\pm
\begin{bmatrix}
b_1 & b_2\\
b_3 & b_4
\end{bmatrix}
=
\begin{bmatrix}
a_1\pm b_1 & a_2\pm b_2\\
a_3\pm b_3 & a_4\pm b_4
\end{bmatrix}
$$

```python
a = np.array([
 [1,2],
 [3,4]])
b = np.array([
 [1,2],
 [3,4]])

print("a + b =\n", a + b)

print("a - b =\n", a - b)
```

#### Producto externo (Hadamard)
$$
\begin{bmatrix}
a_1 & a_2\\
a_3 & a_4
\end{bmatrix}
\odot
\begin{bmatrix}
b_1 & b_2\\
b_3 & b_4
\end{bmatrix}
=
\begin{bmatrix}
a_1\cdot b_1 & a_2\cdot b_2\\
a_3\cdot b_3 & a_4\cdot b_4
\end{bmatrix}
$$

```python
a = np.array([1,2,3])
b = np.array([2,3,4])
print("Producto de vectores a*b =", a * b)

X = np.array([
 [2,3],
 [2,3]])
Y = np.array([
 [3,4],
 [5,6]])
print("Producto de matrices X*Y =\n", X * Y)
```

### Producto punto 
Este tipo de operacin algebraica toma dos vectores y retorna un escalar. Sin embargo, puede aplicarse tambin para matrices, lo que se conoce tambin como producto matricial. La librera _numpy_ usa la funcin _np.dot(A,B)_ para operaciones vectoriales y matriciales. Esta operacin es una de las ms importantes en _deep learning_.

#### Vectorial
$$
\begin{bmatrix}
a_1 & a_2
\end{bmatrix}
\cdot
\begin{bmatrix}
b_1\\
b_2
\end{bmatrix}
=
a_1b_1+a_2b_2
$$

```python
y = np.array([1,2,3])
x = np.array([2,3,4])
print("Producto de punto de vectores x*y =", np.dot(y,x))
```

#### Matricial
En el caso de las matrices, hay que tener en cuenta que no todas las matrices pueden multiplicarse entre s ya que hay un requerimiento para las dimensiones de las matrices: el nmero de columnas de la primer matriz debe ser igual al nmero de filas de la segunda. Como resultado, si multiplicamos una matriz de dimensiones $M\times N$ por otra matriz $N\times K$, obtendremos una matriz de tamao $M\times K$.
$$
\begin{bmatrix}
a_1 & a_2\\
a_3 & a_4
\end{bmatrix}
\cdot
\begin{bmatrix}
b_1 & b_2\\
b_3 & b_4
\end{bmatrix}
=
\begin{bmatrix}
a_1\cdot b_1+a_2\cdot b_3 & a_1\cdot b_2+a_2\cdot b_4\\
a_3\cdot b_1+a_4\cdot b_3 & a_3\cdot b_2+a_4\cdot b_4
\end{bmatrix}
$$

```python
A = np.array([
 [1, 2],
 [3, 4]
 ])
B = np.array([
 [4, 3],
 [2, 1]
 ])
# Multiply
mm = np.dot(A,B)
print(mm)
print(mm.shape)
```

### Transpuesta de una matriz
Las redes neuronales procesan frecuentemente pesos y datos de entrada de diferentes tamaos donde las dimensiones no cumplen los requerimientos de la multiplicacin de matrices. La transposicin de matrices permite rotar una de las matrices (o vectores) de forma que se pueda cumplir con el requerimiento del producto matricial. Para lograr esto, se utiliza la extensin _.T_ al vector o matriz de inters.

```python
b = np.array([
 [1,2,3],
 [4,5,6]
])
print("Dimensiones de b =", b.shape)
b_T = b.T
print("Resultado de b transpuesta =\n", b_T)
print("Dimensiones de b transpuesta =", b_T.shape)
```

### Broadcasting en Numpy
En _numpy_, los requerimientos de operaciones elemento por elemento son facilitadas a travs de un mecanismo llamado _broadcasting_. Mientras que, por ejemplo, para sumar dos matrices se requiere que estas tengan las mismas dimensiones, este mecanismo tambin permite realizar operaciones con matrices de diferentes dimensiones mientras una de ellas tenga una dimensin de tamao 1.

```python
a = np.array([
 [1],
 [2]
])
b = np.array([
 [3,4],
 [5,6]
])
c = np.array([
 [1,2]
])

# Mismo nmero de filas
# Diferente nmero de columnas
# pero a tiene una columna, as que funciona por broadcasting
print(a * b)

# Mismo nmero de columnas
# Diferente nmero de filas
# pero c tiene una fila, as que funciona por broadcasting
print(b * c)

# Diferente nmero de columnas
# Diferente nmero de filas
# pero ambos a y c cumplen la regla de una dimensin
print(a + c)
```
