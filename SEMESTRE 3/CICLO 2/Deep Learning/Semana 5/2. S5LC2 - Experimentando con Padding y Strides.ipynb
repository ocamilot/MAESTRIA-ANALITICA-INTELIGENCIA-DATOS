## Experimentando con Padding y Strides

### Propsito 

En redes neuronales convolucionales se requiere el uso de algunos filtros para facilitar la deteccin de bordes. Uno de estos mtodos es el _padding,_ el cual se refiere a la inclusin de pixeles vacos alrededor de los bordes de una imagen con el fin de conservar su tamao original cuando se aplica un filtro convolucional. Por otra parte, los _strides_ hacen referencia al incremento del tamao del paso en el cual permitimos deslizar un filtro en una imagen.

En este notebook revisaremos el funcionamiento de estos procesos en detalle y los implementaremos para procesar una imagen y evidenciar sus efectos.

### Introduccin

La operacin de convolucin presenta un fenmeno de reduccin de dimensiones que puede llegar a ser problemtico cuando se realizan varias operaciones consecutivas. Revisemos el siguiente ejemplo:

convo_total.png

Ntese que solo hay 16 pasos posibles para que el filtro alcance el final de la imagen tanto de manera vertical como horizontal. Lo anterior lleva a que la imagen de salida sea de dimensiones $4\times 4$. La frmula que utilizamos para calcular el "encogimiento" de la imagen de dimensin $(m\times m)$ teniendo un kernel de tamao $(f\times f)$ corresponde a:
\begin{equation*} 
(m\times m) * (f\times f) = (m-f+1)*(m-f+1) \end{equation*}

Este fenmeno tiene la consecuencia de que si se ejecutan operaciones de convolucin consecutivamente, la imagen final puede ser extremadamente pequea. Este problema de prdida de informacin se debe a que no se puede deslizar el filtro por los pixeles del borde y, por lo tanto, no se pueden hacer las convoluciones de forma completa. Es importante tener en cuenta que esta reduccin de dimensiones se pronuncia cuando tenemos filtros de mayor dimensin.

### Padding

Una forma de solucionar el problema de encogimiento enunciado anteriormente consiste en rellenar nuestra imagen con pixeles vacos (con valor 0) alrededor de los bordes. Esta tcnica se conoce como _padding_.

S4_2_i3.png

Si aplicamos la convolucin con el mismo filtro, esta vez se lograrn hacer 32 pasos y se obtendr una imagen $6\times 6$, del mismo tamao de la imagen original.

S4_2_i4.png

Para lograr aplicar el _padding_ apropiadamente, es necesario aadir suficientes pixeles en los bordes tal que la imagen resultante tenga las mismas dimensiones que la entrada de imagen en la operacin de convolucin. El nmero de pxeles a agregar en cada borde $p$ depende del tamao del filtro $f$ y se puede calcular con la siguiente ecuacin si la dimensin es impar:
\begin{equation} p = \frac{f-1}{2} \end{equation}

Dado que las dimensiones comunes de filtros son de $3\times 3$, $5\times 5$ y $7\times 7$, esta ecuacin suele ser suficiente para aplicar la tcnica satisfactoriamente.

### Strides

Este concepto se refiere al tamao del paso que se da cuando se desliza el filtro convolucional a travs de la imagen de entrada. Por defecto, el paso (_stride_) es de 1 pixel hacia la derecha o hacia abajo. Con un _stride_ de 2, delizaramos el filtro dos pixeles en cada uno de los pasos. 

S4_2_i5.png

Al deslizar el filtro ms rapido, el proceso se terminar ms rpido y por tanto el tamao de la salida ser ms pequeo teniendo en cuenta que este depende del nmero de pasos. Por ejemplo, si deslizamos un filtro de $3\times 3$ en una imagen de $7\times 7$, podemos solo dar dos pasos hasta que se alcance el final de la imagen asumiendo un stride de 2. Por lo anterior, en total hubo 3 pasos en el proceso y por tanto se tendra una salida de $3\times 3$. El resultado de esta convolucin tambin nos permite detectar el borde pero en menos pasos gracias al stride aplicado:

S4_2_i6.png

Podemos calcular el tamao de la salida dependiento de la dimensin del kernel $f$, la dimensin original de la imagen $n$ y el stride $s$:
\begin{equation*} output= \frac{n-f}{s} + 1 \end{equation*}



La justificacin del uso de los _strides_ es que, si bien a menor tamao de paso, ms informacin va a ser capturada en la salida, tambin se tendr ms informacin redundante dado que va a haber sobrelapamiento en los filtros aplicados a la imagen original. Por ejemplo, cuando aplicamos un filtro $3\times 3$ que se mueve con stride de 1, un kernel va a compartir 2/3 de sus pixeles de entrada con el kernel adyacente:

S4_2_i7.png

Por otro lado, si se incrementa el tamao del paso, pocos parmetros se compartiran entre filtros y la salida ser de menor dimensin, lo que lleva a tener en cuenta menor detalle. Adicionalmente, es importante considerar que entre ms operaciones de filtrado se realicen, mayor poder computacional se necesita. Si nuestra red neuronal solo requiere entendimiento de mayor nivel, se puede utilizar un stride ms grande.

### Convolucin usando padding

En primer lugar, se inicializarn los parmetros de entrada: la imagen a analizar, el kernel a aplicar y el nivel de padding deseado. Entonces, vamos a definir un kernel y un valor de padding $pd$:

```python
image = imgTest
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
pd = 2
```

Ahora, necesitamos hacer el cmputo de las dimensiones de la matrz de la salida. Si consideramos padding, se sigue esta frmula: 
$$n_{out}=n_{in}+2pd-k+1$$

Donde $n_{out}$ es la dimensin de salida, $n_{in}$ la de entrada, $p$ el nivel de padding y $k$ es la dimensin del kernel. Lo anterior debe ser implementado en cada dimensin con el objetivo de crear la matriz de salida con el resultado de la convolucin:

```python
KernelDim_x = kernel.shape[0]
KernelDim_y = kernel.shape[1]
ImgDim_x = image.shape[0]
ImgDim_y = image.shape[1]
Output_x = int(ImgDim_x + 2*pd - KernelDim_x + 1)
Output_y = int(ImgDim_y + 2*pd - KernelDim_y + 1)
output = np.zeros((Output_x, Output_y))
```

Ahora, vamos a implementar el padding sobre la imagen original. Para efectos de este laboratorio haremos un padding par, es decir, que se van a agregar el mismo nmero de filas a cada uno de los bordes. En este caso, es posible usar la funcin `Pad` de `NumPy` para crear los bordes de la matriz:

```python
def padding(image, pd):
    padd=np.pad(image, pad_width=pd)
    return (padd)
imagePadded = padding(image,pd)
```

Ahora vamos a continuar con la convolucin de la imagen. En primer lugar, vamos a iterar sobre los pixeles a nivel vertical con un criterio de parada que detenga el ciclo una vez el kernel alcance la parte inferior de la imagen. Luego, se tiene un ciclo que itera sobre cada elemento en x, por lo que aqu tambien vamos a incluir un criterio de parada para cuando estemos llegando al borde derecho de la imagen y se reinicie el proceso de convolucin. 

Finalmente, vamos a incluir la operacin de convolucin que multiplica el filtro por los pixeles de la imagen en cada uno de los desplazamientos, hace la suma de estos productos y los guarda en la matriz de salida. 

```python
#Se itera en cada pixel de la direccin "y"
for y in range(imagePadded.shape[1]):
    #Se revisa si el kernel aplicado sobre a fila "y" no est por fuera de la imagen. 
    #Si lo est, termina el ciclo para "y".
    if y > imagePadded.shape[1] - KernelDim_y:
        break
    #Se itera en cada pixel de la direccin "x"
    for x in range(imagePadded.shape[0]):
        #Se revisa si el kernel aplicado sobre la columna "x" no est por fuera de la imagen.
        #Si lo est, termina el ciclo.
        if x > imagePadded.shape[0] - KernelDim_x:
            break
        #Se aplica el producto de Hadamard y se suman los elementos de la matriz resultante.
        try:
            output[x, y] = (kernel * imagePadded[x: x + KernelDim_x, y: y + KernelDim_y]).sum()
        except:
            break
```

Podemos entonces ver la imagen de salida, as:


```python
plt.figure(dpi=200)
#Se debe reescalar el resultado de Sobel a una imagen en entero que vara de 0 a 255.
convert_back = cv2.convertScaleAbs(output, alpha=255/output.max())
#Para mostrar la imagen correctamente utilizando matplotlib, se deben reorganizar los canales de OpenCV de BGR a RGB.
plt.imshow(cv2.cvtColor(convert_back, cv2.COLOR_BGR2RGB))
plt.show()
```

### Actividad individual: Convolucin incluyendo padding y strides.

Ahora, desarrolle una funcin que reciba la imagen, el kernel, el nivel de padding y, adems, el nivel de stride. Para incluir los strides hay que tener en cuenta varios aspectos:
*   Las dimensiones de la imagen de salida `output` dependen del nivel de strides $s$ como indica la siguiente frmula:
$$n_{out}=\Biggl[\frac{n_{in}+2p-k}{s}\Biggr]+1$$
*   Debe agregar un condicional para cada direccin donde se garantice que la convolucin se realiza segn el tamao de paso indicado por los strides. _Pista:_ Aproveche la operacin mdulo `a%b`, el cual le retorna el residuo de la divisin de a entre b (devuelve 0 cuando a es mltiplo de b, de lo contrario da un entero). 
*   Debido al funcionamiento de la tcnica _stride_, no se realiza convolucin por cada pixel y la dimensin de la imagen de salida se reduce, por lo que debe garantizar que las operaciones se realizan segn el tamao de paso. _Pista:_ Puede utilizar un contador para cada direccin que se vaya actualizando cada vez que se valide la condicin del inciso anterior. El valor de este contador sera el ndice del pixel de la imagen de salida `output`. Tenga en cuenta que, si aplica un nivel de stride de 2, la imagen de salida se encoge a la misma razn para ambas direcciones.

```python
def convolve2D(image, kernel, pd=0, strides=1):
    KernelDim_x = kernel.shape[0]
    KernelDim_y = kernel.shape[1]
    ImgDim_x = image.shape[0]
    ImgDim_y = image.shape[1]

    imagePadded = padding(image,pd)

    # Dimensiones de la imagen con padding
    PaddedDim_x = imagePadded.shape[0]
    PaddedDim_y = imagePadded.shape[1]

    # Clculo de la dimensin de salida considerando padding y stride
    Output_x = ((PaddedDim_x - KernelDim_x) // strides) + 1
    Output_y = ((PaddedDim_y - KernelDim_y) // strides) + 1
    output = np.zeros((Output_x, Output_y))

    # ndices de la imagen de salida
    out_x = 0
    for x in range(0, PaddedDim_x - KernelDim_x + 1, strides):
        out_y = 0
        for y in range(0, PaddedDim_y - KernelDim_y + 1, strides):
            # Producto de Hadamard y suma de la ventana actual
            output[out_x, out_y] = (kernel * imagePadded[x: x + KernelDim_x, y: y + KernelDim_y]).sum()
            out_y += 1
        out_x += 1
    #raise NotImplementedError
    return output
```

```python
# Tests
kernel = np.array([[-1, -1, -1], 
                   [-1, 8, -1], 
                   [-1, -1, -1]])
image = imgTest
pd=2
strides=2
convol = convolve2D(image,kernel,pd,strides)

plt.figure(dpi=100)
#Se debe reescalar el resultado de Sobel a una imagen en entero que vara de 0 a 255.
convert_back = cv2.convertScaleAbs(convol, alpha=255/convol.max())
#Para mostrar la imagen correctamente utilizando matplotlib, se deben reorganizar los canales de OpenCV de BGR a RGB.
plt.imshow(cv2.cvtColor(convert_back, cv2.COLOR_BGR2RGB))
plt.show()

assert round(np.linalg.norm(convol),1) == round(21426.07500220234,1)
```
