### Funciones de activacin

En una red neuronal de tipo feedforward, la informacin se transmite de manera secuencial entre las capas que conforman estos modelos. En particular, el valor de un nodo es el valor que se obtiene de evaluar una funcin de activacin en una transformacin afn de los nodos de la capa anterior. Algebraicamente tendramos:

$$a_{i,j}=f_{i}(\sum_{k=1}^{n_{i-1}}w_{i,k}a_{i-1,k}+b_{i,j})$$

De esta manera, la funcin de activacin puede ser etendida como una manera de resumir la informacin de los nodos anteriores y, por lo tanto, el desempeo del modelo y la convergencia de su estimacin depende en gran medida de la funcin de activacin que se use en cada una de las capas del modelo.


### Funcin de activacin sigmoide

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales entre 0 y 1. Las dos expresiones ms comunes para esta funcin se presentan a continuacin:
$$f(x)=\frac{1}{1+e^{-x}}=\frac{e^{x}}{1+e^{x}}$$
A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model1 = Sequential()
model1.add(Dense(12, input_dim=8, activation='sigmoid'))
model1.add(Dense(8, activation='sigmoid'))
model1.add(Dense(1, activation='sigmoid'))
```

### Funcin de activacin ReLU (rectificadora)

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales mayores o iguales a 0:
$$f(x)=max(0,x)$$
A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def ReLU(x):
    a = []
    for item in x:
        if item<0:
            a.append(0)
        else:
            a.append(item)
    return a
x = np.arange(-10., 10., 0.2)
ReLUplt = ReLU(x)
plt.plot(x,ReLUplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model2 = Sequential()
model2.add(Dense(12, input_dim=8, activation='relu'))
model2.add(Dense(8, activation='relu'))
model2.add(Dense(1, activation='relu'))
```

### Funcin de activacin Softplus

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales mayores o iguales a 0:
$$f(x)=ln(1+e^{x})$$
A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def Softplus(x):
    a = []
    for item in x:
        a.append(np.log(1+math.exp(item)))
    return a
x = np.arange(-10., 10., 0.2)
Softplusplt = Softplus(x)
plt.plot(x,Softplusplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model3 = Sequential()
model3.add(Dense(12, input_dim=8, activation='softplus'))
model3.add(Dense(8, activation='softplus'))
model3.add(Dense(1, activation='softplus'))
```

### Funcin de activacin Softsign

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales entre -1 y 1:
$$f(x)=\frac{x}{1+|x|}$$
A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def Softsign(x):
    a = []
    for item in x:
        a.append(item/(1+abs(item)))
    return a
x = np.arange(-20., 20., 0.2)
Softsignsplt = Softsign(x)
plt.plot(x,Softsignsplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model4 = Sequential()
model4.add(Dense(12, input_dim=8, activation='softsign'))
model4.add(Dense(8, activation='softsign'))
model4.add(Dense(1, activation='softsign'))
```

### Funcin de activacin tangente hiperblica

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales entre -1 y 1:
$$f(x)=tanh(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}=\frac{e^{2x}-1}{e^{2x}+1}$$
A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def tanhHM(x):
    a = []
    for item in x:
        a.append(math.tanh(item))
    return a
x = np.arange(-10., 10., 0.2)
tanhplt = tanhHM(x)
plt.plot(x,tanhplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model5 = Sequential()
model5.add(Dense(12, input_dim=8, activation='tanh'))
model5.add(Dense(8, activation='tanh'))
model5.add(Dense(1, activation='tanh'))
```

### Funcin de activacin SELU

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales mayores o iguales a -1.75809934084737:
$$f(x)=I(x>0)\lambda x+I(x\leq 0)\lambda \alpha(e^{x}-1)$$

Donde $\lambda$ y $\alpha$ son constantes predefinidas, con $\lambda=1.0507009873554804934193349852946$ y $\alpha=1.6732632423543772848170429916717$.

A continuacin se muestra grficamente esta funcin:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def SELU(x):
    a = []
    for item in x:
        if item>0:
            a.append(1.0507009873554804934193349852946*item)
        else:
            a.append(1.0507009873554804934193349852946*1.6732632423543772848170429916717*(math.exp(item)-1))
    return a
x = np.arange(-10., 10., 0.2)
SELUplt = SELU(x)
plt.plot(x,SELUplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model6 = Sequential()
model6.add(Dense(12, input_dim=8, activation='selu'))
model6.add(Dense(8, activation='selu'))
model6.add(Dense(1, activation='selu'))
```

### Funcin de activacin ELU

Esta funcin tiene como dominio los nmeros reales y su rango corresponde a los nmeros reales mayores o iguales a -1:
$$f(x)=I(x>0)x+I(x\leq 0)\alpha(e^{x}-1)$$

Donde $\alpha$ puede ser cualquier nmero real mayor que cero.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def ELU(x,alpha):
    a = []
    for item in x:
        if item>0:
            a.append(item)
        else:
            a.append(alpha*(math.exp(item)-1))
    return a
x = np.arange(-10., 10.,0.2)
ELUplt = ELU(x,3)
plt.plot(x,ELUplt)
plt.show()
```

El cdigo para implementar esta funcin de activacin para las capas de una red neuronal de tipo feedforward se muestra enseguida:

```python
model7 = Sequential()
model7.add(Dense(12, input_dim=8, activation='elu'))
model7.add(Dense(8, activation='elu'))
model7.add(Dense(1, activation='elu'))
```
