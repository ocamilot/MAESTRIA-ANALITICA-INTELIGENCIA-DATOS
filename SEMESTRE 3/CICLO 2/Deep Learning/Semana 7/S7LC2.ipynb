### Definicin de la red

El anlisis de series de tiempo es un problema complejo debido a la dependencia secuencial que existe en las variables de entrada, esto quiere decir que, un dato particular est relacionado con los del pasado de distintas maneras. Estas dependencias pueden ser de largo plazo, lo cual es un impedimento importante para las RNN simples/tradicionales, ya que al tratar de considerar influencias extensas se presenta el problema del gradiente explosivo/desvaneciente. Las redes LSTM logran superar esta limitante utilizando bloques de memoria conectadas en capas a cambio de neuronas. En la siguiente imagen se observa el esquema de un bloque de memoria LSTM:

lstm.png

Un bloque de LSTM posee una serie componentes que lo hacen ms efectivo que una red tradicional. En primer lugar, maneja dos estados que representan la memoria a corto plazo (estado oculto $h_{t-1}$) y la memoria a largo plazo (estado de celda $C_{t-1}$). LSTM se encargar de actualizar ambos estados utilizando el valor de entrada de la secuencia $X_t$ y una serie de puertas (_gates_) que deciden en qu medida una prediccin $h_t$ es afectada por valores anteriores de corto y largo plazo.

En este modelo, los pesos y sesgos entrenados corresponden a los de las puertas, donde cada una de ellas cumple una funcin diferente. Siguiendo el esquema anterior tenemos:
*   **Una puerta del olvido _(Forget Gate)_:** Consiste en un sub-bloque que decide el porcentaje de la informacin de largo plazo que es conservada en el estado de celda a travs de una funcin sigmoide. Note que el resultado de esta puerta multiplica el estado de celda anterior $C_{t-1}$.
*   **Una puerta de entrada _(Input Gate)_:** Los siguientes dos sub-bloques deciden la informacin de entrada y corto plazo que es almacenada en la memoria de largo plazo. Para lograr esto, el primer sub-bloque determina el porcentaje de informacin que se busca recordar utilizando una funcin sigmoide, mientras el segundo sub-bloque enva la informacin a recordar a travs de una funcin tangente hiperblica. Note que los resultados de ambos sub-bloques se multiplican para luego ser sumados a la memoria de largo plazo.
*   **Una puerta de salida _(Output Gate)_:** Los ltimos dos sub-bloques deciden la informacin de largo plazo que ser almacenada como memoria de corto plazo. En este caso, el primer sub-bloque determina el porcentaje de informacin que se busca recordar utilizando una funcin sigmoide y usa la informacin de entrada y la memoria de corto plazo anterior; mientras, el segundo sub-bloque enva la informacin de largo plazo a recordar a travs de una funcin tangente hiperblica. Los resultados de ambos sub-bloques se multiplican para obtener el _output_ del bloque, es decir, la prediccin del siguiente dato. Es importante notar que el ltimo sub-bloque no tiene pesos ni sesgos a entrenar.

Se puede observar en esta arquitectura bsica de LSTM que, para cada uno de los primeros 4 sub-bloques, tenemos dos pesos y un sesgo, lo cual da un total para todo el bloque de 8 pesos y 4 sesgos. ### Implementacin en Keras/Tensorflow

Para interactuar con los diferentes componentes de la red, implementaremos LSTM utilizando la capa que ofrece _Keras_ mediante la aplicacin en un ejemplo. En primer lugar, importaremos las libreras requeridas:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
#Las siguientes dos lneas evitan la generacin de "warnings" y "flags" al importar keras y tensorflow.
#Dichas advertencias no comprometen el funcionamiento del cdigo.
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
#Importacin de keras y tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

Aplicaremos la red sobre una serie de tiempo que representa la generacin de electridad neta mensual medida en billones de kilowatts hora (kWh) en los Estados Unidos (datos tomados de la Administracin de Informacin Energtica de Estados Unidos):

```python
# Se carga el conjunto de datos
dataframe = pd.read_csv('usmelec.csv', usecols=[1])
# Se observan los primeros datos
print(dataframe.head())
# Grfica de la serie
plt.figure(figsize=(10,5))
plt.plot(dataframe)
plt.show()
```

Estos datos tiene una clara tendencia creciente, no slo en el promedio sino en la varianza de la serie. Una prctica comn al entrenar un modelo para prediccin de series de tiempo es estacionalizar los datos. Sin embargo, pondremos a prueba la red LSTM sin realizar esta transformacin de los datos.

Por otro lado, hay una transformacin que s es necesaria para redes LSTM: el escalado de los datos. Debido a que estamos trabajando con funciones sigmoide y tangente hiperblica, es esencial que los datos se reajusten al rango entre 0 y 1:

```python
dataset = dataframe.values
dataset = dataset.astype('float32')
#Se reescala la serie de tiempo
scaler = MinMaxScaler(feature_range=(0, 1))
dataset_rescaled = scaler.fit_transform(dataset)
#Grfica de la serie
plt.figure(figsize=(10,5))
plt.plot(dataset_rescaled)
plt.show()
```

A continuacin, se dividir la serie en una parte inicial para entrenamiento (60%) y una parte final de prueba (40%). Con esta particin, probaremos la capacidad de LSTM para predecir la serie a futuro:

```python
train_percent = 0.6
train_size = int(len(dataset_rescaled) * train_percent)
test_size = len(dataset_rescaled) - train_size
train, test = dataset_rescaled[0:train_size,:], dataset_rescaled[train_size:len(dataset),:]
print("Cantidad de datos para entrenamiento: ",len(train))
print("Cantidad de datos para prueba: ",len(test))
```

El problema de prediccin (_forecast_) consiste en la estimacin del siguiente valor de la serie de tiempo, en este caso, dada la generacin de energa en un mes, se busca predecir el valor del siguiente mes. Para entrenar la red LSTM, debemos convertir el conjunto de datos en un problema de aprendizaje supervisado, donde un dato del pasado busca predecir al siguiente. Para esto, creamos una nueva columna donde los datos son desplazados una posicin (esto es, un rezago o _lag_ igual a 1), con lo cual se busca que la primera columna sea predicha por la segunda. La siguiente funcin permite generar varias columnas dependiendo de la cantidad de lags que se quieran analizar, lo cual nos servir ms adelante:

```python
def timeseries_to_supervised(data, lag=1):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    return df
```

Se aplica la funcin a los sets de entrenamiento y prueba:

```python
lag=1
train_supervised, test_supervised = timeseries_to_supervised(train, lag), timeseries_to_supervised(test, lag)
print(train_supervised)
```

Ntese que es inevitable la generacin de un `NaN` debido a que el primer dato no tiene un dato predecesor con el cual se pueda comparar. Esto se tiene en cuenta en los siguientes cambios de dimensiones de los datos al incluir la variable `lag`:

```python
train_values, test_values = train_supervised.values[lag:,:], test_supervised.values[lag:,:]

train_values = train_values.reshape(train_values.shape[0], train_values.shape[1])
test_values = test_values.reshape(test_values.shape[0], test_values.shape[1])

trainX, trainY = train_values[:, 0:-1], train_values[:, -1]
testX, testY = test_values[:, 0:-1], test_values[:, -1]

# [samples, time steps, features]
trainX = trainX.reshape(trainX.shape[0], lag, 1)
testX = testX.reshape(testX.shape[0], lag, 1)
```

Las sucesivas transformaciones realizadas anteriormente son requeridas por la red LSTM en el siguiente orden: en la primera dimensin estn la cantidad de muestras $(X_{t-1},X_{t})$, en la segunda el nmero de lags, y en la tercera los _features_ a predecir, que en este caso es **un** dato en el futuro.

Antes de establecer los parmetros de la red neuronal, hay que ser conscientes de que los pesos obtenidos luego de entrenar el modelo cambian dependiendo de la inicializacin de los pesos. La consecuencia de esto es que, cada vez que se crea, compila y entrena una red, el desempeo ser diferente en cada ejecucin a pesar de usar los mismos hiperparmetros. Este comportamiento estocstico dificulta la decisin respecto a cules son los hiperparmetros apropiados para solucionar un problema con una red neuronal. Existen dos soluciones para esto:
*   Entrenar varias veces el mismo modelo y guardar una lista con las mtricas de desempeo en test (por ejemplo, RMSE) de cada entrenamiento. Luego, se promedian los resultados y se decide el desempeo del modelo con este resultado (tambin se puede obtener un intervalo de confianza). El problema con esta opcin es que implica mucho tiempo computacional, ya que requiere hacer varios entrenamientos para una sola combinacin de hiperparmetros.
*   Hacer un solo entrenamiento fijando una semilla con la funcin `tf.keras.utils.set_random_seed` justo antes de la creacin de un modelo de red neuronal. Al hacer esto, garantizamos la reproducibilidad, es decir, que los valores de los pesos sern inicializados de la misma manera siempre que se use el mismo valor de semilla, de manera que luego del entrenamiento se obtendrn los mismos pesos. En este caso, el valor de la semilla se puede tratar como un hiperparmetro ms, o se puede fijar y explorar nicamente los hiperparmetros propios de la red. En este notebook, utilizaremos esta ltima opcin para evitar realizar un gran nmero de entrenamientos.

```python
semilla=4269
tf.keras.utils.set_random_seed(semilla)
```

A continuacin, definamos una red bsica LSTM, la cual tendr slo una unidad oculta que se conecta a la neurona de salida con activacin lineal. Se utilizar como funcin de prdida el _Mean Squared Error_, tamao de batch de 1 y un _Early Stopping_ con paciencia de 10 que nos servir en caso que el entrenamiento haya convergido tempranamente:

```python
hidden_units = 1
batch_size = 1
model = Sequential()
model.add(LSTM(hidden_units, batch_input_shape = (batch_size, trainX.shape[1], trainX.shape[2])))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)

model.summary()
```

En el sumario del modelo podemos confirmar que tenemos slo 12 parmetros entrenables en el bloque LSTM: 8 pesos y 4 sesgos. Entrenemos el modelo:

```python
history=model.fit(trainX, trainY, epochs=100, batch_size=batch_size, verbose=2, callbacks=[callback])
```

A modo de ilustracin, observemos los pesos del modelo:

```python
model.get_weights()
```

El primer arreglo corresponde a los pesos relacionados con los inputs/features (cuadro azul en el esquema mostrado a inicios del notebook); el segundo arreglo corresponde a los pesos relacionados con las unidades ocultas (cuadro rojo en el esquema mostrado a inicios del notebook). El tercer arreglo presenta los sesgos de los _gates_.

Observemos la evolucin de la funcin de prdida:

```python
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel('loss')
```

El entrenamiento del modelo convergi, con lo cual podemos realizar predicciones sobre el conjunto de prueba y probar su efectividad con la mtrica RMSE:

```python
# Se realizan predicciones
trainPredict = model.predict(trainX, batch_size=batch_size)
testPredict = model.predict(testX, batch_size=batch_size)
# Se devuelve el escalado de las predicciones
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
# Se calcula RMSE
trainScore = np.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

Para contextualizar el desempeo de la red, grafiquemos la serie original (lnea azul) contra la serie resultante del entrenamiento (lnea naranja) y la predicha con el set de prueba:

```python
# Se ingresan predicciones de train
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[lag:len(trainPredict)+lag, :] = trainPredict
# Se ingresan predicciones de test
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(lag*2):len(dataset), :] = testPredict
# Se grafican las series
plt.figure(figsize=(10,5))
plt.plot(dataset)
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

Se puede observar que, grficamente, la serie predicha sigue la tendencia creciente pero tiene problemas en los valores pico. An podemos mejorar bastante la red en las siguientes secciones, en las que probaremos el cambio de algunos hiperparmetros para hacer la red ms compleja y revisar mejora del desempeo en trminos de RMSE.

### LSTM con varias unidades ocultas

As como en una red neuronal tradicional se pueden agregar ms neuronas a una capa para aumentar el nmero de pesos a entrenar, es posible aumentar el nmero de parmetros de forma equivalente en LSTM. El nmero de total de parmetros depende de la cantidad de _features_ en el input y la cantidad de _hidden units_ establecidas en el modelo. Las _hidden units_ son el equivalente al nmero de neuronas en una capa: ilustremos cmo cambiara el nmero de pesos en el siguiente esquema con 2 unidades ocultas:

lstm2.png

Note que al aumentar las _hidden units_, no slo aumenta la dimensin de los vectores $h$ y $C$, sino que tambin lo hacen las neuronas de cada _gate_. Como resultado, como el vector concatenado tiene dimensin 3 (1 feature + 2 hidden units) y dos neuronas por puerta, se tiene un total de 6 pesos por _gate_. A esto se suma un sesgo por neurona, lo cual nos da un total de 8 parmetros por puerta y un total $8\times 4 = 32$ parmetros entrenables para esta nueva red LSTM. Generalizando, se puede decir que para $x$ features y $h$ hidden units, el total de parmetros es igual a $4\times ((x+h)\times h+h)$.

Evaluemos esta red con la serie de datos de aplicacin:

```python
# Preparacin de datos
trainX, trainY = train_values[:, 0:-1], train_values[:, -1]
testX, testY = test_values[:, 0:-1], test_values[:, -1]
# [samples, time steps, features]
trainX = trainX.reshape(trainX.shape[0], lag, 1)
testX = testX.reshape(testX.shape[0], lag, 1)

#Se fija la semilla
tf.keras.utils.set_random_seed(semilla)
#Modelo LSTM
hidden_units = 2
batch_size = 1
model = Sequential()
model.add(LSTM(hidden_units, batch_input_shape = (batch_size, trainX.shape[1], trainX.shape[2])))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)

model.summary()
```

Podemos verificar que el nmero de parmetros entrenables en LSTM es 32. Ahora realicemos el entrenamiento:

```python
history=model.fit(trainX, trainY, epochs=100, batch_size=batch_size, verbose=2, callbacks=[callback])
```

```python
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel('loss')
```

```python
# Se realizan predicciones
trainPredict = model.predict(trainX, batch_size=batch_size)
testPredict = model.predict(testX, batch_size=batch_size)
# Se devuelve el escalado de las predicciones
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
# Se calcula RMSE
trainScore = np.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
# Se ingresan predicciones de train
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[lag:len(trainPredict)+lag, :] = trainPredict
# Se ingresan predicciones de test
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(lag*2):len(dataset), :] = testPredict
# Se grafican las series
plt.figure(figsize=(10,5))
plt.plot(dataset)
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

Compare el RMSE de este modelo con el anterior, mejor?. Es importante recordar que el resultado del entrenamiento vara debido a la estocasticidad inherente del problema, por lo que es posible que obtenga un resultado a veces mejor, a veces peor. En nuestro caso, depender del valor de la semilla.

### Uso de pasos de tiempo en redes LSTM

Hasta este momento, hemos utilizado LSTM considerando un problema de aprendizaje supervisado con un rezago (lag) igual a 1. Sin embargo, es posible tener en cuenta dependencias temporales mayores aumentando el valor del lag utilizando **pasos de tiempo** (`timesteps`). Por ejemplo, dado que tenemos datos mensuales, se puede predecir lo que suceder en un mes (_feature_ igual a 1) teniendo en cuenta siempre el ao anterior (ltimos 12 meses), y as identificar caractersticas peridicas de la serie de tiempo. Generemos una base de datos con lag 12 utilizando la funcin `timeseries_to_supervised`:

```python
lag=12
train_supervised, test_supervised = timeseries_to_supervised(train, lag), timeseries_to_supervised(test, lag)
print(train_supervised)
```

Se puede observar que se generan varios valores `NaN` debido a que estamos buscando que cada dato tenga 12 datos previos para analizar, por lo que habr una reduccin del dataset de train en este sentido.

```python
### Preparacin de datos
train_values, test_values = train_supervised.values[lag:,:], test_supervised.values[lag:,:]

train_values = train_values.reshape(train_values.shape[0], train_values.shape[1])
test_values = test_values.reshape(test_values.shape[0], test_values.shape[1])

trainX, trainY = train_values[:, 0:-1], train_values[:, -1]
testX, testY = test_values[:, 0:-1], test_values[:, -1]

# [samples, time steps, features]
trainX = trainX.reshape(trainX.shape[0], lag, 1)
testX = testX.reshape(testX.shape[0], lag, 1)
```

Ahora construyamos el modelo. Tenga en cuenta que por como est construido el cdigo, la dimensin de `trainX` ya le indica a _Keras_ el nmero de pasos de tiempo.

```python
#Se fija la semilla
tf.keras.utils.set_random_seed(semilla)
hidden_units = 2
batch_size = 1
model = Sequential()
model.add(LSTM(hidden_units, batch_input_shape = (batch_size, trainX.shape[1], trainX.shape[2])))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)

model.summary()
```

Observe que el nmero de parmetros no ha cambiado con respecto al caso anterior de lag igual a 1. Esto sucede porque, a pesar de tener ms pasos de tiempo, el entrenamiento sigue siendo recursivo y se comparten los mismos pesos y sesgos en todas las iteraciones de cada lag. Esto permite un ahorro de memoria significativo, dejando que los parmetros slo dependan de las unidades ocultas y las features que se buscan predecir.

Entrenemos el modelo y detallemos si hay diferencias en el desempeo:

```python
history=model.fit(trainX, trainY, epochs=100, batch_size=batch_size, verbose=2, callbacks=[callback])
```

```python
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel('loss')
```

```python
# Se realizan predicciones
trainPredict = model.predict(trainX, batch_size=batch_size)
testPredict = model.predict(testX, batch_size=batch_size)
# Se devuelve el escalado de las predicciones
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
# Se calcula RMSE
trainScore = np.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
# Se ingresan predicciones de train
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[lag:len(trainPredict)+lag, :] = trainPredict
# Se ingresan predicciones de test
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(lag*2):len(dataset), :] = testPredict
# Se grafican las series
plt.figure(figsize=(10,5))
plt.plot(dataset)
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

El desempeo en trminos de RMSE es considerablemente superior, lo cual se confirma grficamente. Note cmo la implementacin de pasos de tiempo puede aumentar el desempeo del modelo considerablemente sin necesidad del entrenamiento de un gran nmero de parmetros.

```python
def timeseries_to_supervised(data, lag=1):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    return df


def miLSTM(semilla,lag,hidden_units,batch_size,optimizador,paciencia,epocas):
    # your code here
    # Carga de datos
    dataframe = pd.read_csv('AAPL.csv', usecols=[1])
    dataset = dataframe.values.astype('float32')

    # Escalado
    scaler = MinMaxScaler(feature_range=(0, 1))
    dataset_rescaled = scaler.fit_transform(dataset)

    # Divisin train/test
    train_size = int(len(dataset_rescaled) * 0.6)
    train, test = dataset_rescaled[:train_size], dataset_rescaled[train_size:]

    # Transformacin supervisada
    train_supervised = timeseries_to_supervised(train, lag)
    test_supervised = timeseries_to_supervised(test, lag)

    train_values = train_supervised.values[lag:, :]
    test_values = test_supervised.values[lag:, :]

    trainX, trainY = train_values[:, :-1], train_values[:, -1]
    testX, testY = test_values[:, :-1], test_values[:, -1]

    trainX = trainX.reshape(trainX.shape[0], lag, 1)
    testX = testX.reshape(testX.shape[0], lag, 1)

    # Fijar semilla
    tf.keras.utils.set_random_seed(semilla)

    # Crear modelo
    model = Sequential()
    model.add(LSTM(hidden_units, batch_input_shape=(batch_size, lag, 1)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer=optimizador)

    # Early stopping
    callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=paciencia, restore_best_weights=True)

    # Entrenar modelo
    history = model.fit(trainX, trainY, epochs=epocas, batch_size=batch_size, verbose=2, callbacks=[callback])

    
    return trainX, trainY, testX, testY, model, history
```
